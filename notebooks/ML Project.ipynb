```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("MiningProcess_Flotation_Plant_Database.csv", decimal=",")
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df = df.groupby('date').agg(['mean', 'std'])
```

```python
df.columns = ['_'.join(col) for col in df.columns]
```

```python
df = df.sort_index()
```

```python
airflow_columns = ["Flotation Column 01 Air Flow_mean", "Flotation Column 02 Air Flow_mean", "Flotation Column 03 Air Flow_mean", "Flotation Column 04 Air Flow_mean", "Flotation Column 05 Air Flow_mean", "Flotation Column 06 Air Flow_mean", "Flotation Column 07 Air Flow_mean"]
```

```python
airflow_columns_1 = ['Flotation Column 01 Air Flow_std', 'Flotation Column 02 Air Flow_std',
       'Flotation Column 03 Air Flow_std', 'Flotation Column 04 Air Flow_std',
       'Flotation Column 05 Air Flow_std', 'Flotation Column 06 Air Flow_std',
       'Flotation Column 07 Air Flow_std']
```

```python
level_columns = ["Flotation Column 01 Level_mean", "Flotation Column 02 Level_mean", "Flotation Column 03 Level_mean", "Flotation Column 04 Level_mean", "Flotation Column 05 Level_mean", "Flotation Column 06 Level_mean", "Flotation Column 07 Level_mean"]
```

```python
level_columns_1 = ['Flotation Column 01 Level_std', 'Flotation Column 02 Level_std',
       'Flotation Column 03 Level_std', 'Flotation Column 04 Level_std',
       'Flotation Column 05 Level_std', 'Flotation Column 06 Level_std',
       'Flotation Column 07 Level_std']
```

```python
df["airflow_mean"] = df[airflow_columns].mean(axis=1)
```

```python
df["airflow_std"] = df[airflow_columns_1].std(axis=1)
```

```python
df["level_mean"] = df[level_columns].mean(axis=1)
```

```python
df["level_std"] = df[level_columns_1].std(axis=1)
```

```python
df.drop(level_columns, axis=1, inplace=True)
```

```python
df.drop(airflow_columns, axis=1, inplace=True)
```

```python
variables_to_lag = ['% Iron Feed_mean', '% Silica Feed_mean', 'Starch Flow_mean', 'Amina Flow_mean', 'Ore Pulp pH_mean']
```

```python
for col in variables_to_lag:
    df[f'{col}_lag1'] = df[col].shift(1)
    df[f'{col}_lag3'] = df[col].shift(3)
```

```python
for col in variables_to_lag:
    df[f'{col}_mean_3h'] = df[col].rolling(window=3).mean()
    df[f'{col}_std_3h'] = df[col].rolling(window=3).std()
```

```python
df = df.dropna()
```

```python
df.drop(airflow_columns_1, axis=1, inplace=True)
```

```python
df.drop(level_columns_1, axis=1, inplace=True)
```
